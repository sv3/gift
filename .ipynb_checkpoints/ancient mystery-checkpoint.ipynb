```python
import Image
import inline_display
import numpy as np
```

```python
i = Image.open('fire.jpg')
```

```python
r = i.copy()
r.thumbnail((500,500))
```

```python
pix = np.array(r)
h, w, _ = pix.shape

avg = np.mean(pix, axis=2)
row_avg = np.mean(avg, axis=0)

pix_mod = pix

for y in range(h):
    maxx = row_avg[y]
    pix_mod[y] += maxx*0.5
    
np.n
```

```python
mod = Image.fromarray(pix_mod)
mod.convert('P', palette=Image.ADAPTIVE, colors=60)
```

```python
from skimage import io, color
lab = color.rgb2lab(r)
rgb = color.lab2rgb(lab)
```

```python
Image.fromarray(rgb.astype('int'))
```
